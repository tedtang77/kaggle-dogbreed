## 5. Validate predictions

```python
val_batches, probs = vgg.test(valid_path, batch_size)
```

```python
filenames = val_batches.filenames
# get class indices of all classes eg. 'affenpinscher', 'afghan_hound', 'african_hunting_dog', 'airedale'
classes = vgg.classes  
# fact
expected_labels = val_batches.classes #classes (breeds) no. 0~119  
expected_label_indices = np.array([classes[int(expected_labels[i])] for i in range(len(expected_labels))])
# predict
our_prediction_idxs = np.argmax(probs, axis=1)
our_prediction_probs = np.array([probs[i, our_prediction_idxs[i]] for i in range(len(our_prediction_idxs))])
our_prediction_labels = our_prediction_idxs
our_prediction_label_indices = np.array([classes[idx] for idx in our_prediction_idxs])
```

```python
from keras.preprocessing import image

def plots_labels(ims, figsize=(12,6), rows=1, interp=False, expected_labels=None, our_labels=None, titles=None):
    if type(ims[0]) is np.ndarray:
        ims = np.array(ims).astype(np.uint8)
        if (ims.shape[-1] != 3):
            ims = ims.transpose((0,2,3,1))
    f = plt.figure(figsize=figsize)
    cols = len(ims)//rows if len(ims) % 2 == 0 else len(ims)//rows + 1
    for i in range(len(ims)):
        sp = f.add_subplot(rows, cols, i+1)
        sp.axis('Off')
        if expected_labels is not None and our_labels is not None:
            sp.set_title('F: '+expected_labels[i]+'\nP: '+our_labels[i], fontsize=16)
        elif titles is not None:
            sp.set_title(titles[i], fontsize=16)
        plt.imshow(ims[i], interpolation=None if interp else 'none')

#Helper function to plot images by index in the validation set 
#Plots is a helper function in utils.py
def plots_idx(idx, expected_labels=None, our_labels=None, titles=None):
    plots_labels([image.load_img(valid_path + filenames[i]) for i in idx], 
                 expected_labels=expected_labels, our_labels=our_labels,
                titles=titles)

#Number of images to view for each visualization task
n_view = 4
```

```python
#1. A few correct labels at random
correct = np.squeeze(np.where(our_prediction_labels==expected_labels))
print("Found %d correct labels" % len(correct))
idx = permutation(correct)[:n_view]
if len(correct):
    plots_idx(idx, expected_label_indices[idx], our_prediction_label_indices[idx])
```

```python
#2 A few incorrect labels at random
incorrect = np.squeeze(np.where(our_prediction_labels!=expected_labels))
print("Found %d incorrect labels" % len(incorrect))
idx = permutation(incorrect)[:n_view]
if len(incorrect):
    plots_idx(idx, expected_label_indices[idx], our_prediction_label_indices[idx])
```

```python
#3. The images we most confident were any dog breed, and are actually this dog breed
correct = np.squeeze(np.where(our_prediction_labels==expected_labels))
#print("Found %d confident correct dog breed labels" % len(correct))
most_correct = np.argsort(our_prediction_probs[correct])[::-1][:n_view]
if len(most_correct):
    plots_idx(correct[most_correct], expected_label_indices[correct[most_correct]], our_prediction_label_indices[correct[most_correct]])
```

```python
#4. The images we were most confident, but are actually wrong
incorrect = np.squeeze(np.where(our_prediction_labels!=expected_labels))
most_incorrect = np.argsort(our_prediction_probs[incorrect])[::-1][:n_view]
if len(most_incorrect):
    plots_idx(incorrect[most_incorrect], expected_label_indices[incorrect[most_incorrect]], our_prediction_label_indices[incorrect[most_incorrect]])
```

```python
#5. The most uncertain labels (ie those with probability closest to 0.5).
most_uncertain = np.argsort(our_prediction_probs)[:n_view]
if len(most_uncertain):
    plots_idx(most_uncertain, expected_label_indices[most_uncertain], our_prediction_label_indices[most_uncertain])
```
