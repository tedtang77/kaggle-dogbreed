#### Read label.csv into correct data structure (... again for repeating finetuning and training)

````python
%cd $DATA_HOME_DIR
labels = pd.read_csv('labels.csv', header=0,  delimiter=",")
#print(len(labels))
labels = labels.sort_values(by='id', ascending=True)
labels.set_index("id", drop=True, inplace=True)
labels_dict = labels.to_dict(orient="index")
#print(labels_dict['001513dfcb2ffafc82cccf4d8bbaba97']['breed'])

breed_list = list(labels.groupby('breed').count().index)
#print(len(breed_list))
#print(breed_list[:10])
````

####  finetuning and training

````python
%cd $DATA_HOME_DIR

#Set path to sample/ path if desired
path = DATA_HOME_DIR + '/' # + '/sample/' # 
test_path = path + '/test/' #We use all the test data
results_path = path + '/results/'
train_path = path + '/train/'
valid_path = path + '/valid/'
model_path = path + '/models/'
if not os.path.exists(model_path): os.mkdir(model_path)
````

````python
#import Vgg16 helper class
vgg = Vgg16()
#model = vgg.model
````

````python
#Set constants. You can experiment with no_of_epochs to improve the model
batch_size=64
no_of_epochs=10
````

````python
#Finetune the model
batches = vgg.get_batches(train_path, batch_size=batch_size)
val_batches = vgg.get_batches(valid_path, batch_size=batch_size*2)
vgg.finetune(batches)

#Not sure if we set this for all fits
vgg.model.optimizer.lr = 0.001
````

````python
%cd $results_path
#Notice we are passing in the validation dataset to the fit() method
#For each epoch we test our model against the validation set
latest_weights_filename = None

#latest_weights_filename = 'ft_test.h5' 
#vgg.model.save_weights(latest_weights_filename)

for epoch in range(no_of_epochs):
    print("Running epoch: %d" % epoch)
    vgg.fit(batches, val_batches, nb_epoch=1)
    latest_weights_filename = 'ft%d.h5' % epoch
    vgg.model.save_weights(latest_weights_filename)
print("Completed %s fit operations" % no_of_epochs)
````

## 5. Validate Predictions

Keras' *fit()* function conveniently shows us the value of the loss function, and the accuracy, after every epoch ("*epoch*" refers to one full run through all training examples). The most important metrics for us to look at are for the validation set, since we want to check for over-fitting. 

- **Tip**: with our first model we should try to overfit before we start worrying about how to reduce over-fitting - there's no point even thinking about regularization, data augmentation, etc if you're still under-fitting! (We'll be looking at these techniques shortly).

As well as looking at the overall metrics, it's also a good idea to look at examples of each of:
1. A few correct labels at random
2. A few incorrect labels at random
3. The most correct labels of each class (ie those with highest probability that are correct)
4. The most incorrect labels of each class (ie those with highest probability that are incorrect)
5. The most uncertain labels (ie those with probability closest to 0.5).

Let's see what we can learn from these examples. (In general, this is a particularly useful technique for debugging problems in the model. However, since this model is so simple, there may not be too much to learn at this stage.)

Calculate predictions on validation set, so we can find correct and incorrect examples:

````python
%cd $results_path
latest_weights_filename = 'ft2.h5'
vgg.model.load_weights(results_path+latest_weights_filename)
#vgg.model.load_weights(latest_weights_filename)
````

````python
val_batches, probs = vgg.test(valid_path, batch_size = batch_size)
````

````python
filenames = val_batches.filenames
expected_labels = val_batches.classes #breeds no. 0~119  
expected_labels_breeds = np.array([breed_list[int(expected_labels[i])] for i in range(len(expected_labels))])

#Choose the max probability to generate labels
our_predictions = np.argmax(probs, axis=1)
our_labels = our_predictions
our_labels_breeds = np.array([breed_list[int(our_labels[i])] for i in range(len(our_labels))])
````

````python
assert(len(expected_labels)==len(our_labels))
assert(len(our_labels)==len(expected_labels_breeds))
````

````python
from keras.preprocessing import image

def plots_labels(ims, figsize=(12,6), rows=1, interp=False, expected_labels=None, our_labels=None, titles=None):
    if type(ims[0]) is np.ndarray:
        ims = np.array(ims).astype(np.uint8)
        if (ims.shape[-1] != 3):
            ims = ims.transpose((0,2,3,1))
    f = plt.figure(figsize=figsize)
    cols = len(ims)//rows if len(ims) % 2 == 0 else len(ims)//rows + 1
    for i in range(len(ims)):
        sp = f.add_subplot(rows, cols, i+1)
        sp.axis('Off')
        if expected_labels is not None and our_labels is not None:
            sp.set_title('F: '+expected_labels[i]+'\nP: '+our_labels[i], fontsize=16)
        elif titles is not None:
            sp.set_title(titles[i], fontsize=16)
        plt.imshow(ims[i], interpolation=None if interp else 'none')

#Helper function to plot images by index in the validation set 
#Plots is a helper function in utils.py
def plots_idx(idx, expected_labels=None, our_labels=None, titles=None):
    plots_labels([image.load_img(valid_path + filenames[i]) for i in idx], 
                 expected_labels=expected_labels, our_labels=our_labels,
                titles=titles)

#Number of images to view for each visualization task
n_view = 4
````

````python
#1. A few correct labels at random
correct = np.where(our_labels==expected_labels)[0]
print("Found %d correct labels" % len(correct))
idx = permutation(correct)[:n_view]
if len(correct):
    plots_idx(idx, expected_labels_breeds[idx], our_labels_breeds[idx])
````

````python
#2. A few incorrect labels at random
incorrect = np.where(our_labels!=expected_labels)[0]
print("Found %d incorrect labels" % len(incorrect))
idx = permutation(incorrect)[:n_view]
if len(incorrect):
    plots_idx(idx, expected_labels_breeds[idx], our_labels_breeds[idx])
````

````python
#3. The images we most confident were any dog breed, and are actually any dog breed
correct = np.where(our_labels==expected_labels)[0]
print("Found %d confident correct dog breed labels" % len(correct))
most_correct = np.argsort(our_predictions[correct])[::-1][:n_view]
if len(most_correct):
    plots_idx(correct[most_correct], expected_labels_breeds[correct][most_correct], our_labels_breeds[correct][most_correct])
````

````python
#4a. The images we were most confident, but are actually wrong
incorrect = np.where(our_labels!=expected_labels)[0]
print("Found %d incorrect labels" % len(incorrect))
if len(incorrect):
    most_incorrect = np.argsort(our_predictions[incorrect])[::-1][:n_view]
    plots_idx(incorrect[most_incorrect], expected_labels_breeds[incorrect][most_incorrect], our_labels_breeds[incorrect][most_incorrect])
````

````python
#5. The most uncertain labels (ie those with probability closest to 0.5).
most_uncertain = np.argsort(np.max(probs, axis=1))[:n_view]
if len(most_uncertain):
    plots_idx(most_uncertain, titles=np.max(probs, axis=1)[most_uncertain])
````
